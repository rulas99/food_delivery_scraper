```python
import pandas as pd
from glob import glob
from tqdm import tqdm
```

```python
uber_files = glob('./results/uber_new/*.json')
uber_data = pd.concat([pd.read_json(f, lines=False) for f in tqdm(uber_files)],
                      ignore_index=True)

rappi_files = glob('./results/rappi_new/*.json')
rappi_data = pd.concat([pd.read_json(f, lines=False) for f in tqdm(rappi_files)],
                       ignore_index=True)
```

```python
complete_data = pd.concat([uber_data, rappi_data], ignore_index=True)
complete_data.to_csv('./results/complete_data_new.csv', index=False)
complete_data
```
